```python
from sklearn.cluster import KMeans                # Import all necessary modules
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("ECG_featcher1.csv")              # Importing feature extracted from mit bih arrhythmia dataset
df.head()
```

```python
km = KMeans(n_clusters=4)                                               # Using ML model without scaling
patient = df["patient"]
df1 = df.drop(["patient"],axis=1)
df1
```

```python
y_predicted = km.fit_predict(df1)                 # ML models predited data (without scaling)
y_predicted
```

```python
col_names = ["BPM","time_of_QR_(sec)","time_of_RS_(sec)","time_of_QRS(sec)"]   # Scaling all features using MinMaxScaler
features = df1[col_names]

scaler = MinMaxScaler().fit(features.values)
features = scaler.transform(features.values)

df1[col_names] = features
df1
```

```python
km = KMeans(n_clusters=4)                                                       # Setting k = 4 for final ML model
```

```python
y_predicted = km.fit_predict(df1)                                             # Fitting and predicting using scaled data
y_predicted
df2 = df1.copy()
df2["cluter"] = y_predicted
cluter = df2.copy()
df2                                                                            # Final Classification in four classes
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df1)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel("K")                                                   # Elbow method for selecting K, which is about 4 in this case
plt.ylabel("Sum of Squared error")
plt.plot(k_rng,sse)
```

```python
x=df2.iloc[:,:-1]
y=df2.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
from sklearn.tree import DecisionTreeClassifier
tc=DecisionTreeClassifier(criterion='gini')
tc.fit(x_train,y_train)
tc.score(x_test,y_test)
```
